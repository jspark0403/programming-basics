## 📌 예제 1 : 자기소개 출력하기

### 문제
**이름**, **학번**, **전공**을 입력 하여,  아래와 같은 형식으로 자기소개를 출력하는 프로그램을 작성하세요.

**출력 예시:**
```
안녕하세요! 저는 홍길동입니다.
학번: 20251234
전공: 기계공학과
잘 부탁드립니다! 😊
```

---

### 💡 힌트
- `input("질문 내용: ")` 을 사용하면 사용자가 값을 입력할 수 있어요.
- 입력받은 값은 변수에 저장해두세요. 예: `name = input("이름: ")`
- `print()` 안에서 문자열과 변수를 합치려면 **f-string** 을 사용하세요.
  - 예: `print(f"안녕하세요! 저는 {name}입니다.")`
- 변수 3개가 필요: 이름, 학번, 전공

````python
# ✏️ 예제 1 - 여기에 코드를 작성하세요!

# 1. 이름을 입력 받기
name = input("이름: ")

# 2. 학번을 입력 받기
stdnum = input("학번: ")

# 3. 전공을 입력 받기
major = input("전공: ")

# 4. 자기소개 출력하기
print(f"안녕하세요! 저는 {name}입니다.\n학번: {stdnum}\n전공: {major}\n잘 부탁드립니다! 😊")

print('name: ' + name)
# +연산자도 알아두어야 함.
````

## 📌 예제 2 : 두 숫자의 합과 곱 출력하기

### 문제
**두 개의 숫자**를 입력 받아서, 그 두 숫자의 **합**과 **곱**을 출력하는 프로그램을 작성하세요.

**출력 예시 (입력: 3, 5):**
```
첫 번째 숫자: 3
두 번째 숫자: 5
두 수의 합: 8
두 수의 곱: 15
```

---

### 💡 힌트
- `input()` 으로 받은 값은 기본적으로 **문자열(str)** 이에요.
- 숫자 계산을 하려면 **정수형(int)** 으로 변환하세요.
  - 예: `num1 = int(input("첫 번째 숫자: "))`
- 합은 `+` 연산자, 곱은 `*` 연산자를 사용하세요.

````python
# ✏️ 예제 2 - 여기에 코드를 작성하세요!

# 1. 첫 번째 숫자 입력 받기 (정수로 변환!)
a = int(input("첫 번째 숫자: "))

# 2. 두 번째 숫자 입력 받기 (정수로 변환!)
b = int(input("두 번째 숫자: "))

# 3. 합 계산하기
sum = a+b

# 4. 곱 계산하기
mul = a*b

# 5. 결과 출력하기
print(f'''첫 번째 숫자: {a}
두 번째 숫자: {b}
두 수의 합: {sum}
두 수의 곱: {mul}''')
````

## 📌 예제 3 : 전체 초(sec)를 몇 분 몇 초로 변환하기

### 문제
**전체 초(sec)** 를 입력 하여,  그것이 **몇 분 몇 초** 인지 계산해서 출력하는 프로그램을 작성하세요.

**출력 예시 (입력: 145):**
```
입력한 시간: 145초
2분 25초
```

---

### 💡 힌트
- **분(min)** 을 구하려면: 전체 초 ÷ 60의 **몫** → `//` 연산자 사용!
  - 예: `145 // 60` → `2`
- **초(sec)** 를 구하려면: 전체 초 ÷ 60의 **나머지** → `%` 연산자 사용!
  - 예: `145 % 60` → `25`
- `//` : 나눗셈의 **몫** (소수점 버림)
- `%`  : 나눗셈의 **나머지**

````python
# ✏️ 예제 3 - 여기에 코드를 작성하세요!

# 1. 전체 초(sec) 입력 받기 (정수로 변환!)
time = int(input('입력한 시간: '))

# 2. 몇 분인지 계산하기 (힌트: // 연산자)
min = 145//60

# 3. 나머지 몇 초인지 계산하기 (힌트: % 연산자)
sec = 145%60

# 4. 결과 출력하기
print(f'''
입력한시간: {time}초
{min}분 {sec}초
''')
````
